```python
# è‡ªå‹•è»Šï¼ˆCarï¼‰ã‚¯ãƒ©ã‚¹ã®å®šç¾©
class Car:
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
    name: strï¼ˆè»Šç¨®ï¼‰ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ï¼špublic
    distance: intï¼ˆèµ°è¡Œè·é›¢ï½‹ï½ï¼‰ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ï¼špublic
  '''
 

  # ç™ºé€²ã®è¡¨ç¤ºï¼šdepartureãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šdeparture
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼š'ç™ºé€²ã—ã¾ã™ï¼ï¼'ã¨è¡¨ç¤º
    æˆ»ã‚Šå€¤ï¼šãªã—
  '''


  # é€²ã‚“ã è·é›¢ã®è¡¨ç¤ºï¼šrunningãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šrunning
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼šç¾åœ¨ä½•ï½‹ï½é€²ã‚“ã ã‹ã‚’è¡¨ç¤º
    æˆ»ã‚Šå€¤ï¼šãªã—
  '''
 

  # å‰é€²ã®è¨ˆç®—ã¨è¡¨ç¤ºï¼šgoãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šgo
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼š'ï¼‘ï½‹ï½å‰é€²ã—ã¾ã™ï¼ï¼'ã¨è¡¨ç¤ºã—ã€ãƒ•ã‚£ãƒ¼ãƒ«ãƒ‰distanceã‚’ã‚¤ãƒ³ã‚¯ãƒªãƒ¡ãƒ³ãƒˆ
  '''


  # æ›´æ–°ã®è¨ˆç®—ã¨è¡¨ç¤ºï¼šbackãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šback
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼š'ï¼‘ï½‹ï½å¾Œé€²ã—ã¾ã™ï¼ï¼'ã¨è¡¨ç¤ºã—ã€ãƒ•ã‚£ãƒ¼ãƒ«ãƒ‰distanceã‚’ãƒ‡ã‚¯ãƒªãƒ¡ãƒ³ãƒˆ
  '''
```

### â—†äººé–“ã‚¯ãƒ©ã‚¹ã‚’ä½œæˆã—ã€å„ç¨®å¥åº·é–¢é€£ãƒ¡ã‚½ãƒƒãƒ‰ã‚’å®šç¾©ã€å€‹äººã®å¥åº·æƒ…å ±ã‚’è¡¨ç¤ºã€‚ã¾ãŸã€ã‚²ãƒƒã‚¿ãƒ¼ãƒ»ã‚»ãƒƒã‚¿ãƒ¼ã®å®šç¾©ã‚‚è¡Œã†

```python
# äººé–“ã‚¯ãƒ©ã‚¹ï¼ˆPersonï¼‰ã®å®šç¾©
class Person:
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
    __name: æ°åï¼ˆå‹ str ã‚¹ã‚³ãƒ¼ãƒ— privateï¼‰
    __age: å¹´é½¢ï¼ˆå‹ int ã‚¹ã‚³ãƒ¼ãƒ— privateï¼‰
    __gender: æ€§åˆ¥ï¼ˆå‹ str ã‚¹ã‚³ãƒ¼ãƒ— privateï¼‰
    __height: èº«é•·ï¼ˆå‹ float ã‚¹ã‚³ãƒ¼ãƒ— privateï¼‰
    __weight: ä½“é‡ï¼ˆå‹ float ã‚¹ã‚³ãƒ¼ãƒ— privateï¼‰
  '''
  def __init__(self, name: str, age: int, 
      gender: str, height: float, weight: float) -> None:
    self.__name = name
    self.__age = age
    self.__gender = gender
    self.__height = height
    self.__weight = weight

  # å„ãƒ•ã‚£ãƒ¼ãƒ«ãƒ‰ã®ã‚²ãƒƒã‚¿ãƒ¼å®šç¾©
  # nameã‚²ãƒƒã‚¿ãƒ¼
  @property
  def name(self) -> str:
    return self.__name

  # ageã‚²ãƒƒã‚¿ãƒ¼
  @property
  def age(self) -> int:
    return self.__age

  # genderã‚²ãƒƒã‚¿ãƒ¼
  @property
  def gender(self) -> str:
    return self.__gender

  # heightã‚²ãƒƒã‚¿ãƒ¼
  @property
  def height(self) -> float:
    return self.__height

  # weightã‚²ãƒƒã‚¿ãƒ¼
  @property
  def weight(self) -> float:
    return self.__weight

  # å„ãƒ•ã‚£ãƒ¼ãƒ«ãƒ‰ã®ã‚»ãƒƒã‚¿ãƒ¼å®šç¾©
  # nameã‚»ãƒƒã‚¿ãƒ¼
  @name.setter
  def name(self, name: str) -> None:
    self.__name = name

  # ageã‚»ãƒƒã‚¿ãƒ¼
  # ageã«ï¼ä»¥ä¸‹ã®å€¤ãŒè¨­å®šã•ã‚ŒãŸæ™‚ã¯ï¼ã‚’è¨­å®šã™ã‚‹
  @age.setter
  def age(self, age: int) -> None:
    self.__age = age if age > 0 else 0

  # heightã‚»ãƒƒã‚¿ãƒ¼
  # heightã«ï¼ï¼ï¼ä»¥ä¸‹ã®å€¤ãŒè¨­å®šã•ã‚ŒãŸæ™‚ã¯ï¼ï¼ï¼ã‚’è¨­å®šã™ã‚‹
  @height.setter
  def height(self, height: float) -> None:
    self.__height = height if height > 0.0 else 0.0

  # weightã‚»ãƒƒã‚¿ãƒ¼
  # weightã«ï¼ï¼ï¼ä»¥ä¸‹ã®å€¤ãŒè¨­å®šã•ã‚ŒãŸæ™‚ã¯ï¼ï¼ï¼ã‚’è¨­å®šã™ã‚‹
  @weight.setter
  def weight(self, weight: float) -> None:
    self.__weight = weight if weight > 0.0 else 0.0


  # å€‹äººæƒ…å ±ã‚’è¡¨ç¤ºã™ã‚‹ãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šdisp_info
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼š
      æ°åã€å¹´é½¢ã€æ€§åˆ¥ã€èº«é•·ã€ä½“é‡ã‚’è¡¨ç¤º
    æˆ»ã‚Šå€¤ï¼šãªã—
  '''
  def disp_info(self) -> None:
    print('æ°åï¼š{}'.format(self.__name))
    print('å¹´é½¢ï¼š{} æ­³'.format(self.__age))
    print('æ€§åˆ¥ï¼š{}'.format(self.__gender))
    print('èº«é•·ï¼š{} cm'.format(self.__height))
    print('ä½“é‡ï¼š{} kg'.format(self.__weight))

  # BMIå€¤ã‚’è¨ˆç®—ã™ã‚‹ãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šget_bmi
      å¼•æ•°ï¼šãªã—
      å‡¦ç†ï¼š
        èº«é•·ï¼ˆï½å˜ä½ï¼‰ã¨ä½“é‡ï¼ˆï½‹ï½‡å˜ä½ï¼‰ã‹ã‚‰BMIå€¤ã‚’è¨ˆç®—ã—è¿”ã™
        è¨ˆç®—å¼ã¯ä»¥ä¸‹ã®é€šã‚Š
          BMIå€¤ = ä½“é‡ï¼ˆï½‹ï½‡å˜ä½ï¼‰ï¼ã€€èº«é•·ï¼ˆï½å˜ä½ï¼‰ã®ï¼’ä¹—
      æˆ»ã‚Šå€¤ï¼šBMIå€¤ï¼ˆå‹ floatï¼‰
  '''
  def get_bmi(self) -> float:
    # èº«é•·ï¼ˆï½ƒï½å˜ä½ï¼‰ã‚’ï½å˜ä½ã«æ›ç®—
    height_m = self.__height / 100

    # BMIå€¤ã‚’è¨ˆç®—ã—è¿”ã™
    return self.__weight / (height_m * height_m)

  # é©æ­£ä½“é‡ã‚’è¨ˆç®—ã—è‚¥æº€åº¦ï¼ˆå‹ strï¼‰ã‚’è¿”ã™
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šget_obestity
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼š
      BMIå€¤ã‚’ã‚‚ã¨ã‚ãã®å€¤ã«ã‚ˆã‚Šè‚¥æº€åº¦ã‚’åˆ¤æ–­ã—
      è‚¥æº€åº¦ã‚’ç¤ºã™æ–‡å­—åˆ—ã‚’è¿”ã™
      è‚¥æº€åº¦ã‚’ç¤ºã™æ–‡å­—åˆ—ã¯ä»¥ä¸‹ã®é€šã‚Š
        BMIå€¤ãŒ
          ï¼‘ï¼˜ï¼ï¼•æœªæº€ãªã‚‰ã€€'ä½ä½“é‡ï¼ˆã‚„ã›å‹ï¼‰'
          ï¼‘ï¼˜ï¼ï¼•ä»¥ä¸Šã€ï¼’ï¼•æœªæº€ãªã‚‰ 'æ™®é€šä½“é‡'
          ï¼’ï¼•ä»¥ä¸Šã€ï¼“ï¼æœªæº€ãªã‚‰ 'è‚¥æº€ï¼ˆï¼‘åº¦ï¼‰'
          ï¼“ï¼ä»¥ä¸Šã€ï¼“ï¼•æœªæº€ãªã‚‰ 'è‚¥æº€ï¼ˆï¼’åº¦ï¼‰'
          ï¼“ï¼•ä»¥ä¸Šã€ï¼”ï¼æœªæº€ãªã‚‰ 'è‚¥æº€ï¼ˆï¼“åº¦ï¼‰'
          ï¼”ï¼ä»¥ä¸Šãªã‚‰ 'è‚¥æº€ï¼ˆï¼”åº¦ï¼‰'
    æˆ»ã‚Šå€¤ï¼šè‚¥æº€åº¦ã‚’è¡¨ã™æ–‡å­—åˆ—ï¼ˆå‹ strï¼‰
  '''
  def get_obesity(self) -> str:
    # get_bmiãƒ¡ã‚½ãƒƒãƒ‰ã‚’å‘¼ã³å‡ºã—BMIå€¤ã‚’å¤‰æ•°bmiã«ä»£å…¥
    bmi = self.get_bmi()

    # BMIå€¤ã‚’åˆ¤æ–­ã—ã¦è‚¥æº€åº¦ã‚’è¡¨ã™æ–‡å­—åˆ—ã‚’è¿”ã™
    if bmi < 18.5:
      return 'ä½ä½“é‡ï¼ˆã‚„ã›å‹ï¼‰'
    elif bmi < 25:
      return 'æ™®é€šä½“é‡'
    elif bmi < 30:
      return 'è‚¥æº€ï¼ˆï¼‘åº¦ï¼‰'
    elif bmi < 35:
      return 'è‚¥æº€ï¼ˆï¼’åº¦ï¼‰'
    elif bmi < 40:
      return 'è‚¥æº€ï¼ˆï¼“åº¦ï¼‰'
    else:
      return 'è‚¥æº€ï¼ˆï¼”åº¦ï¼‰'

  # èº«é•·ã‹ã‚‰é©æ­£ä½“é‡ã‚’è¨ˆç®—ã—è¿”ã™ãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šsuitable_weight
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼š
      èº«é•·ï¼ˆï½å˜ä½ï¼‰ã‹ã‚‰é©æ­£ä½“é‡ï¼ˆï½‹ï½‡å˜ä½ï¼‰ã‚’è¨ˆç®—ã—è¿”ã™
      é©æ­£ä½“é‡ã¯ä»¥ä¸‹ã®è¨ˆç®—å¼ã«å¾“ã†
        é©æ­£ä½“é‡ = èº«é•·ï¼ˆï½å˜ä½ï¼‰ã®ï¼’ä¹— x ï¼’ï¼’
    æˆ»ã‚Šå€¤ï¼šé©æ­£ä½“é‡ï¼ˆå‹ floatï¼‰
  '''
  def suitable_weight(self) -> float:
    # èº«é•·ï¼ˆï½ƒï½å˜ä½ï¼‰ã‚’ï½å˜ä½ã«æ›ç®—ã™ã‚‹
    height_m = self.__height / 100

    # é©æ­£ä½“é‡ã‚’è¨ˆç®—ã—è¿”ã™
    return height_m * height_m * 22
```

```python
# Personã‚¯ãƒ©ã‚¹ã®å®šç¾©
class Person:
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  _name: æ°åï¼ˆå‹ str ã‚¹ã‚³ãƒ¼ãƒ— protectedï¼‰
  _job: è·æ¥­ï¼ˆå‹ str ã‚¹ã‚³ãƒ¼ãƒ— protectedï¼‰
  '''
  def __init__(self, name: str, job: str) -> None:
    self._name = name
    self._job = job

  # è‡ªå·±ç´¹ä»‹ã™ã‚‹ãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©ï¼ˆæ°åã€è·æ¥­ã‚’è¡¨ç¤ºï¼‰
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šintroduce
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼šæ°åã¨å¹´é½¢ã‚’è¡¨ç¤º
    æˆ»ã‚Šå€¤ï¼šãªã—
  '''
  def introduce(self) -> None:
    # æ°åã‚’è¡¨ç¤º
    print('æ°åï¼š{}'.format(self._name))
    # è·æ¥­ã‚’è¡¨ç¤º
    print('è·æ¥­ï¼š{}'.format(self._job))
```

```python
# Teacherï¼ˆå…ˆç”Ÿï¼‰ã‚¯ãƒ©ã‚¹ã®å®šç¾©ã€€Personã‚¯ãƒ©ã‚¹ã‚’ç¶™æ‰¿
class Teacher(Person):
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  _name, _job: ã‚¹ãƒ¼ãƒ‘ãƒ¼ã‚¯ãƒ©ã‚¹ã®ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿ã‚’å‘¼ã³å‡ºã™
  _subject: æ‹…å½“ç§‘ç›®ï¼ˆå‹ str ã‚¹ã‚³ãƒ¼ãƒ— protectedï¼‰
  '''
  def __init__(self, name: str, job: str, subject: str) -> None:
    super().__init__(name, job)
    self._subject = subject

  # è‡ªå·±ç´¹ä»‹ãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©ï¼ˆã‚ªãƒ¼ãƒãƒ¼ãƒ©ã‚¤ãƒ‰ï¼‰
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šintroduce
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼š
      ã‚¹ãƒ¼ãƒ‘ãƒ¼ã‚¯ãƒ©ã‚¹ã®introduceãƒ¡ã‚½ãƒƒãƒ‰ã‚’å‘¼ã³å‡ºã—æ°åã¨å¹´é½¢ã‚’è¡¨ç¤º
      æ‹…å½“ç§‘ç›®ã‚’è¡¨ç¤º
    æˆ»ã‚Šå€¤ï¼šãªã—
  '''
  def introduce(self) -> None:
    super().introduce()
    print('æ‹…å½“ç§‘ç›®ï¼š{}'.format(self._subject))
```

```python
# Bookï¼ˆæ›¸ç±ï¼‰ã‚¯ãƒ©ã‚¹ã®å®šç¾©
class Book:
  # ã‚¯ãƒ©ã‚¹å¤‰æ•°__counterã‚’ï¼ã«åˆæœŸåŒ–
  # ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ãŒç”Ÿæˆã•ã‚Œã‚‹ãŸã³ã«ç•ªå·ã‚’ã‚¤ãƒ³ã‚¯ãƒªãƒ¡ãƒ³ãƒˆã™ã‚‹
  __counter = 0

  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  _title: æ›¸åï¼ˆå‹ strã€€ã‚¹ã‚³ãƒ¼ãƒ— protectedï¼‰
  _author: è‘—è€…ï¼ˆå‹ strã€€ã‚¹ã‚³ãƒ¼ãƒ— protectedï¼‰
  _price: ä¾¡æ ¼ï¼ˆå‹ intã€€ã‚¹ã‚³ãƒ¼ãƒ— protectedï¼‰
  '''
  def __init__(self, title: str, author: str, price: int) -> None:
    # æ›¸åã®è¨­å®š
    self._title = title
    # è‘—è€…ã®è¨­å®š
    self._author = author
    # ä¾¡æ ¼ã®è¨­å®š
    self._price = price
    # ã‚¯ãƒ©ã‚¹å¤‰æ•°__counterã®ã‚¤ãƒ³ã‚¯ãƒªãƒ¡ãƒ³ãƒˆ
    Book.__counter += 1
    # è­˜åˆ¥ç•ªå·ã®è¨­å®šï¼ˆã‚¯ãƒ©ã‚¹å¤‰æ•°__counterã‚’è¨­å®šï¼‰
    self._id = Book.__counter

  # æ›¸ç±æƒ…å ±ã‚’è¡¨ç¤ºã™ã‚‹ãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰ç¨®åˆ¥ï¼šã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ãƒ¡ã‚½ãƒƒãƒ‰
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šdisplay_info
      å¼•æ•°ï¼šãªã—
      å‡¦ç†ï¼šæ›¸åã€è‘—è€…åã€ä¾¡æ ¼ã€è­˜åˆ¥ç•ªå·ã‚’è¡¨ç¤ºã™ã‚‹
      æˆ»ã‚Šå€¤ï¼šãªã—
  '''
  def display_info(self) -> None:
    print('æ›¸åï¼š{}'.format(self._title))
    print('è‘—è€…åï¼š{}'.format(self._author))
    print('ä¾¡æ ¼ï¼š{} å††'.format(self._price))
    print('è­˜åˆ¥ç•ªå·ï¼š{}'.format(self._id))

  # ä¸ãˆãŸè­˜åˆ¥ç•ªå·ã®æœ€å¤§å€¤ï¼ˆæœ€å¾Œã«ä¸ãˆãŸè­˜åˆ¥ç•ªå·ï¼‰ã‚’è¿”ã™ã‚¯ãƒ©ã‚¹ãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰ç¨®åˆ¥ï¼šã‚¯ãƒ©ã‚¹ãƒ¡ã‚½ãƒƒãƒ‰
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šmax_counter
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼šæœ€å¾Œã«ä¸ãˆãŸè­˜åˆ¥ç•ªå·ã‚’è¿”ã™
    æˆ»ã‚Šå€¤ï¼šè­˜åˆ¥ç•ªå·ï¼ˆå‹ intï¼‰
  '''
  @classmethod
  def max_counter(cls) -> int:
    return cls.__counter
```

```python
# PaperBookï¼ˆç´™åª’ä½“æ›¸ç±ï¼‰ã‚¯ãƒ©ã‚¹ã®å®šç¾©ï¼ˆBookã‚¯ãƒ©ã‚¹ã‚’ç¶™æ‰¿ï¼‰
class PaperBook(Book):
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  title, author, price: ã‚¹ãƒ¼ãƒ‘ãƒ¼ã‚¯ãƒ©ã‚¹ã®ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿ã‚’å‘¼ã³å‡ºã™ 
  page_number: ãƒšãƒ¼ã‚¸æ•°ï¼ˆå‹ intã€€ã‚¹ã‚³ãƒ¼ãƒ— protectedï¼‰
  '''
  def __init__(self, title: str, author: str, 
          price: int, page_number: int) -> None:
    super().__init__(title, author, price)
    self._page_number = page_number

  # æ›¸ç±æƒ…å ±ã‚’è¡¨ç¤ºã™ã‚‹ãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰ç¨®åˆ¥ï¼šã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ãƒ¡ã‚½ãƒƒãƒ‰
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šdisplay_info
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼šæ›¸åã€è‘—è€…åã€ä¾¡æ ¼ã€è­˜åˆ¥ç•ªå·ã€ãƒšãƒ¼ã‚¸æ•°ã‚’è¡¨ç¤ºã™ã‚‹
  '''
  def display_info(self) -> None:
    super().display_info()
    print('ãƒšãƒ¼ã‚¸æ•°ï¼š{} ãƒšãƒ¼ã‚¸'.format(self._page_number))
```

```python
# EBookï¼ˆé›»å­æ›¸ç±ï¼‰ã‚¯ãƒ©ã‚¹ã®å®šç¾©ï¼ˆBookã‚¯ãƒ©ã‚¹ã‚’ç¶™æ‰¿ï¼‰
class EBook(Book):
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  title, author, price: ã‚¹ãƒ¼ãƒ‘ãƒ¼ã‚¯ãƒ©ã‚¹ã®ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿ã‚’å‘¼ã³å‡ºã™ 
  file_size: ãƒ•ã‚¡ã‚¤ãƒ«ã‚µã‚¤ã‚ºï¼ˆå‹ intã€€ã‚¹ã‚³ãƒ¼ãƒ— protectedï¼‰
  '''
  def __init__(self, title: str, author: str, 
          price: int, file_size: int) -> None:
    super().__init__(title, author, price)
    self._file_size = file_size

  # æ›¸ç±æƒ…å ±ã‚’è¡¨ç¤ºã™ã‚‹ãƒ¡ã‚½ãƒƒãƒ‰ã®å®šç¾©
  '''
  ãƒ¡ã‚½ãƒƒãƒ‰ã‚·ã‚°ãƒ‹ãƒãƒ£
    ãƒ¡ã‚½ãƒƒãƒ‰ç¨®åˆ¥ï¼šã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ãƒ¡ã‚½ãƒƒãƒ‰
    ãƒ¡ã‚½ãƒƒãƒ‰åï¼šdisplay_info
    å¼•æ•°ï¼šãªã—
    å‡¦ç†ï¼šæ›¸åã€è‘—è€…åã€ä¾¡æ ¼ã€è­˜åˆ¥ç•ªå·ã€ãƒ•ã‚¡ã‚¤ãƒ«ã‚µã‚¤ã‚ºã‚’è¡¨ç¤ºã™ã‚‹
  '''
  def display_info(self) -> None:
    super().display_info()
    print('ãƒ•ã‚¡ã‚¤ãƒ«ã‚µã‚¤ã‚ºï¼š{} KB'.format(self._file_size))
```

```python
# Personã‚¯ãƒ©ã‚¹ã®å®šç¾©
class Person:
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  name: æ°åï¼ˆå‹ strã€€ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ protectedï¼‰
  age: å¹´é½¢ï¼ˆå‹ int  ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ protectedï¼‰
  '''
  def __init__(self, name: str, age: int) -> None:
    self._name = name
    self._age = age
```

```python
# InterfaceGreetã‚¯ãƒ©ã‚¹ã®å®šç¾©ï¼ˆã‚¤ãƒ³ã‚¿ãƒ•ã‚§ãƒ¼ã‚¹çš„ã«ä½¿ç”¨ï¼‰
class InterfaceGreet:
  # ãƒ¡ã‚½ãƒƒãƒ‰å®šç¾©ï¼šæŒ¨æ‹¶ã™ã‚‹
  def greeting(self) -> None:
    pass
```

```python
# InterfaceMealã‚¯ãƒ©ã‚¹ã®å®šç¾©ï¼ˆã‚¤ãƒ³ã‚¿ãƒ•ã‚§ãƒ¼ã‚¹çš„ã«ä½¿ç”¨ï¼‰
class InterfaceMeal:
  # ãƒ¡ã‚½ãƒƒãƒ‰å®šç¾©ï¼šé£Ÿäº‹æƒ…å ±è¡¨ç¤º
  def meal(self) -> None:
    pass
```

```python
# Japaneaseã‚¯ãƒ©ã‚¹ã®å®šç¾©
# Person, InterfaceGreet, InterfaceMealã‚’å¤šé‡ç¶™æ‰¿
class Japanease(Person, InterfaceGreet, InterfaceMeal):
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  _name, _age: ã‚¹ãƒ¼ãƒ‘ãƒ¼ã‚¯ãƒ©ã‚¹ã®ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿ã‚’å‘¼ã³å‡ºã—
  _country: å›½ç±ï¼ˆå‹ str  ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ï¼šprotectedï¼‰
  '''
  def __init__(self, name: str, age: int, country: str) -> None:
    super().__init__(name, age)
    self._country = country

  # è‡ªå·±ç´¹ä»‹ãƒ¡ã‚½ãƒƒãƒ‰å®šç¾©ï¼šInterfaceGreetã®ãƒ¡ã‚½ãƒƒãƒ‰ã‚’å®Ÿè£…
  def greeting(self) -> None:
    print('ã“ã‚“ã«ã¡ã¯ã€ç§ã¯{}ã§ã™ã€‚å¹´é½¢ã¯{}æ­³ã§ã™ã€‚'
        .format(self._name, self._age))
    print('å›½ç±ã¯ã€{} ã§ã™ã€‚'.format(self._country))

  # é£Ÿäº‹ç´¹ä»‹ãƒ¡ã‚½ãƒƒãƒ‰å®šç¾©ï¼šInterfaceMealã®ãƒ¡ã‚½ãƒƒãƒ‰ã‚’å®Ÿè£…
  def meal(self) -> None:
    print('{}äººã¯ä¸»ã«ç±³ã¨é­šä»‹é¡ã‚’é£Ÿã¹ã¾ã™ã€‚'.format(self._country))
```

```python
# Americanã‚¯ãƒ©ã‚¹ã®å®šç¾©
# Person, InterfaceGreet, InterfaceMealã‚’å¤šé‡ç¶™æ‰¿
class American(Person, InterfaceGreet, InterfaceMeal):
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  _name, _age: ã‚¹ãƒ¼ãƒ‘ãƒ¼ã‚¯ãƒ©ã‚¹ã®ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿ã‚’å‘¼ã³å‡ºã—
  _country: å›½ç±ï¼ˆå‹ str  ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ï¼šprotectedï¼‰
  '''
  def __init__(self, name: str, age: int, country: str) -> None:
    super().__init__(name, age)
    self._country = country

  # è‡ªå·±ç´¹ä»‹ãƒ¡ã‚½ãƒƒãƒ‰å®šç¾©ï¼šInterfaceGreetã®ãƒ¡ã‚½ãƒƒãƒ‰ã‚’å®Ÿè£…
  def greeting(self) -> None:
    print('Hello, I am {}. I am {} years old.'
        .format(self._name, self._age))
    print('Nationality is {}.'.format(self._country))

  # é£Ÿäº‹ç´¹ä»‹ãƒ¡ã‚½ãƒƒãƒ‰å®šç¾©ï¼šInterfaceMealã®ãƒ¡ã‚½ãƒƒãƒ‰ã‚’å®Ÿè£…
  def meal(self) -> None:
    print('{}s mainly eat meat and bread.'.format(self._country))
```

```python
# Japaneaseã‚¯ãƒ©ã‚¹ã®ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã‚’ç”Ÿæˆ
japanease = Japanease('åœŸæ–¹æ­³ä¸‰', 38, 'æ—¥æœ¬')
# Americanã‚¯ãƒ©ã‚¹ã®ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ã‚’ç”Ÿæˆ
american = American('John F Kennedy', 42, 'American')
```

```python
# æŒ¨æ‹¶
japanease.greeting()
# é£Ÿäº‹
japanease.meal()
```

```python
# abcãƒ‘ãƒƒã‚±ãƒ¼ã‚¸ã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
from abc import ABCMeta
from abc import abstractmethod
```

```python
# Personã‚¯ãƒ©ã‚¹ã®å®šç¾©
class Person:
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  name: æ°åï¼ˆå‹ strã€€ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ protectedï¼‰
  age: å¹´é½¢ï¼ˆå‹ int  ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ protectedï¼‰
  '''
  def __init__(self, name: str, age: int) -> None:
    self._name = name
    self._age = age
```

```python
# InterfaceGreetæŠ½è±¡ã‚¯ãƒ©ã‚¹ã®å®šç¾©
class InterfaceGreet(metaclass = ABCMeta):
  # æŠ½è±¡ãƒ¡ã‚½ãƒƒãƒ‰greetingã®å®šç¾©
  @abstractmethod
  def greeting(self) -> None:
    pass
```

```python
# InterfaceMealæŠ½è±¡ã‚¯ãƒ©ã‚¹ã®å®šç¾©
class InterfaceMeal(metaclass = ABCMeta):
  # æŠ½è±¡ãƒ¡ã‚½ãƒƒãƒ‰mealã®å®šç¾©
  @abstractmethod
  def meal(self) -> None:
    pass
```

```python
# Japaneaseã‚¯ãƒ©ã‚¹ã®å®šç¾©
# Person, InterfaceGreet, InterfaceMealã‚’å¤šé‡ç¶™æ‰¿
class Japanease(Person, InterfaceGreet, InterfaceMeal):
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  _name, _age: ã‚¹ãƒ¼ãƒ‘ãƒ¼ã‚¯ãƒ©ã‚¹ã®ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿ã‚’å‘¼ã³å‡ºã—
  _country: å›½ç±ï¼ˆå‹ str  ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ï¼šprotectedï¼‰
  '''
  def __init__(self, name: str, age: int, country: str) -> None:
    super().__init__(name, age)
    self._country = country

  # æŠ½è±¡ãƒ¡ã‚½ãƒƒãƒ‰greetingã®å®Ÿè£…
  def greeting(self) -> None:
    print('ã“ã‚“ã«ã¡ã¯ã€ç§ã¯{}ã§ã™ã€‚å¹´é½¢ã¯{}æ­³ã§ã™ã€‚'
        .format(self._name, self._age))
    print('å›½ç±ã¯ã€{} ã§ã™ã€‚'.format(self._country))

  # æŠ½è±¡ãƒ¡ã‚½ãƒƒãƒ‰mealã®å®Ÿè£…
  def meal(self) -> None:
    print('{}äººã¯ä¸»ã«ç±³ã¨é­šä»‹é¡ã‚’é£Ÿã¹ã¾ã™ã€‚'.format(self._country))
```

```python
# Americanã‚¯ãƒ©ã‚¹ã®å®šç¾©
# Person, InterfaceGreet, InterfaceMealã‚’å¤šé‡ç¶™æ‰¿
class American(Person, InterfaceGreet, InterfaceMeal):
  # ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿å®šç¾©
  '''
  _name, _age: ã‚¹ãƒ¼ãƒ‘ãƒ¼ã‚¯ãƒ©ã‚¹ã®ã‚³ãƒ³ã‚¹ãƒˆãƒ©ã‚¯ã‚¿ã‚’å‘¼ã³å‡ºã—
  _country: å›½ç±ï¼ˆå‹ str  ã‚¢ã‚¯ã‚»ã‚¹å±æ€§ï¼šprotectedï¼‰
  '''
  def __init__(self, name: str, age: int, country: str) -> None:
    super().__init__(name, age)
    self._country = country

  # æŠ½è±¡ãƒ¡ã‚½ãƒƒãƒ‰greetingã®å®Ÿè£…
  def greeting(self) -> None:
    print('Hello, I am {}. I am {} years old.'
        .format(self._name, self._age))
    print('Nationality is {}.'.format(self._country))

  # æŠ½è±¡ãƒ¡ã‚½ãƒƒãƒ‰mealã®å®Ÿè£…
  def meal(self) -> None:
    print('{}s mainly eat meat and bread.'.format(self._country))
```

```python
# æŒ¨æ‹¶
japanease.greeting()
# é£Ÿäº‹
japanease.meal()
```

```python
# æŒ¨æ‹¶
american.greeting()
# é£Ÿäº‹
american.meal()
```
